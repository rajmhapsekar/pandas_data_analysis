```python
### classify movies based on rating(excellent,good,average)
```

```python
def rating(rating):
    if rating>=7.0:
        return 'Excellent'
    elif rating>=6.0:
        return 'good'
    else:
        return 'avg'
```

```python
df['rat_cat']=df['Rating'].apply(rating)
```
